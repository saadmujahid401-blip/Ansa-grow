### **Question 36**
(Visualization) 'Drama' aur 'Action' genres ki movies ke 'Gross' revenue ko compare karne ke liye ek Grouped Box Plot ya Violin Plot banayein.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("movie.csv")

# Remove missing values
df = df.dropna(subset=['Genre', 'Gross'])

# Convert Gross into numeric
df['Gross'] = (
    df['Gross']
    .astype(str)
    .str.replace(',', '', regex=False)
    .astype(float)
)

# Select Drama and Action movies
genre_data = df[
    df['Genre'].str.contains('Drama|Action', case=False, na=False)
].copy()

# Create separate Gross lists
drama_gross = genre_data[
    genre_data['Genre'].str.contains('Drama', case=False)
]['Gross']

action_gross = genre_data[
    genre_data['Genre'].str.contains('Action', case=False)
]['Gross']

# Box Plot
plt.figure(figsize=(8,5))

plt.boxplot(
    [drama_gross, action_gross],
    labels=['Drama', 'Action']
)

plt.title("Gross Revenue Comparison: Drama vs Action")
plt.xlabel("Genre")
plt.ylabel("Gross Revenue")

plt.grid(axis='y', alpha=0.3)

plt.show()
```

### **Question 37**

```python
import pandas as pd

# Load dataset
df = pd.read_csv("movie.csv")

# Remove missing values
df = df.dropna(subset=['Director', 'No_of_Votes'])

# Convert No_of_Votes into numeric
df['No_of_Votes'] = (
    df['No_of_Votes']
    .astype(str)
    .str.replace(',', '', regex=False)
    .astype(int)
)

# Filter movies with votes greater than 500,000
high_votes = df[df['No_of_Votes'] > 500000]

# Count movies per Director
director_count = (
    high_votes.groupby('Director')
    .size()
    .reset_index(name='Movie_Count')
    .sort_values(by='Movie_Count', ascending=False)
)

# Display result
print(director_count)
```

### **Question 40**
(Visualization) Top 5 highest rated movies aur unke 'Meta_score' vs 'IMDB_Rating' (scaled to 100) ka ek Side-by-Side Comparison Bar Chart banayein.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv("movie.csv")

# Remove missing values
df = df.dropna(subset=['Series_Title', 'IMDB_Rating', 'Meta_score'])

# Top 5 highest rated movies
top5 = df.sort_values(
    by='IMDB_Rating',
    ascending=False
).head(5)

# Scale IMDB_Rating from 10 to 100
top5['IMDB_Scaled'] = top5['IMDB_Rating'] * 10

# Prepare bar positions
x = np.arange(len(top5))
width = 0.35

# Side-by-side Bar Chart
plt.figure(figsize=(10,5))

plt.bar(
    x - width/2,
    top5['Meta_score'],
    width,
    label='Meta Score'
)

plt.bar(
    x + width/2,
    top5['IMDB_Scaled'],
    width,
    label='IMDB Rating (Scaled /100)'
)

plt.xticks(
    x,
    top5['Series_Title'],
    rotation=45,
    ha='right'
)

plt.title("Top 5 Highest Rated Movies: Meta Score vs IMDB Rating")
plt.xlabel("Movies")
plt.ylabel("Score (Out of 100)")

plt.legend()

plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
