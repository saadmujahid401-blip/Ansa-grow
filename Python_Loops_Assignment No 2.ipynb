# **Q12:** User se ek number lein aur check karein ke wo **Prime Number** hai ya nahi (using loops).

````python
# type your code here
num = int(input("Enter a number"))

if num > 1:
    prime = True

    for i in range (2, num):
        if num % i == 0 :
            prime = False
            break
            
        if prime:
            print (num,"is a prime number")
        else:
            print (num,"is not a prime number")
else:
    print (num,"is not a prime number")
        
````

# **Q18:** Ek manual counting karein ke ek string (e.g., 'Learning Python is fun') mein kitne **Vowels** (a, e, i, o, u) hain.

````python
# type your code here
string = ("Learning Python is fun")
vowels = ("aeiou")
count = 0

for char in string:
    if char in vowels:
        count += 1
print("Total vowels:", count)
````

# **Q20:** Ek 'Number Guessing Game' banayein jahan program ek number (1-20) fix kare aur user ko 3 chances de guess karne ke liye. `while loop` aur `break` ka use karein.

````python
# type your code here
secret_number = 17
chances = 3

while chances > 0:
    guess = int(input("Guess a number:"))

    if guess == secret_number:
        print("Congratulation number is this")
        break
    else:
        chances -= 1
        print("wrong guess")

        if chances > 0:
            print("you have left",chances)

if chances == 0 :
    print("game over,this is the secret number",secret_number)
            
````
