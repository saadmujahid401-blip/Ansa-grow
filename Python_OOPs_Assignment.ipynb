**Q1:** Aik `Student` class banayein jismein `name` aur `roll_number` private variables (`__`) hon. In variables ko access aur modify karne ke liye getter aur setter methods likhein.

```python
class Student:
    # Constructor
    def __init__(self, name, roll_number):
        self.__name = name              # Private variable
        self.__roll_number = roll_number # Private variable

    # Getter method
    def get_name(self):
        return self.__name

    def get_roll_number(self):
        return self.__roll_number

    # Setter method
    def set_name(self, name):
        self.__name = name

    def set_roll_number(self, roll_number):
        self.__roll_number = roll_number


# Creating object
s1 = Student("Ali", 101)

# Accessing private variables using getter
print("Student Name:", s1.get_name())
print("Roll Number:", s1.get_roll_number())

# Modifying private variables using setter
s1.set_name("Ahmed")
s1.set_roll_number(202)

# Display updated values
print("\nAfter Modification:")
print("Student Name:", s1.get_name())
print("Roll Number:", s1.get_roll_number())
```

**Q2:** Aik `BankAccount` class banayein jismein `__balance` private variable ho. Aik `deposit()` aur aik `withdraw()` method banayein, check karein ke withdraw karte waqt balance zero se kam na ho.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   

    # Deposit method
    def deposit(self, amount):
        self.__balance += amount
        print(amount, "deposit ho gaye.")
        print("Current Balance:", self.__balance)

    # Withdraw method
    def withdraw(self, amount):
        if self.__balance - amount >= 0:
            self.__balance -= amount
            print(amount, "withdraw ho gaye.")
            print("Current Balance:", self.__balance)
        else:
            print("Error: Balance zero se kam nahi ho sakta.")

    
    def check_balance(self):
        return self.__balance



account = BankAccount(4000)

account.deposit(1500)

account.withdraw(4000)

account.withdraw(2000)
```

**Q3:** Aik class banayein `Employee` jismein salary private ho. Agar koi salary ko minus (negative) value set karne ki koshish kare, toh setter method 'Invalid Salary' ka error message show kare.

```python
class Employee:
    def __init__(self, salary):
        self.__salary = salary   # private variable

    # Setter method
    def set_salary(self, salary):
        if salary >= 0:
            self.__salary = salary
            print("Salary updated successfully.")
        else:
            print("Invalid Salary")

    # Getter method
    def get_salary(self):
        return self.__salary


# Object creation
emp = Employee(60000)

# Current salary
print("Salary:", emp.get_salary())

# Valid salary update
emp.set_salary(80000)

# Negative salary attempt
emp.set_salary(-20000)

print("Final Salary:", emp.get_salary())
```

**Q9:** Child class ke constructor (`__init__`) ke andar parent class ke constructor ko call karne ke liye `super()` ka use kaise hota hai? Aik practical example likhein.

```python
# Parent Class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


# Child Class
class Student(Person):
    def __init__(self, name, age, roll_number):
        # Parent class constructor call
        super().__init__(name, age)

        # Child class attribute
        self.roll_number = roll_number


# Object creation
s1 = Student("Ali", 20, 101)

# Printing details
print("Name:", s1.name)
print("Age:", s1.age)
print("Roll Number:", s1.roll_number)
```

**Q10:** Hierarchical Inheritance kya hoti hai? Aik parent class `Shape` banayein aur us se do alag child classes `Circle` aur `Square` banakar samjhayein.

```python
# Write your code here for Q10
# Parent Class
class Shape:
    def area(self):
        print("Every shape has an area")


# Child Class 1
class Circle(Shape):
    def circle_area(self):
        print("Circle area = π × r × r")


# Child Class 2
class Square(Shape):
    def square_area(self):
        print("Square area = side × side")


# Circle object
c1 = Circle()
c1.area()          # Parent class method
c1.circle_area()   # Circle class method


# Square object
s1 = Square()
s1.area()          # Parent class method
s1.square_area()   # Square class method
```

**Q11:** Method Overriding kya hai? Parent class `Bird` banayein jisme `fly()` method ho jo print kare 'Most birds can fly'. Child class `Ostrich` banayein jo is method ko override kare aur print kare 'Ostriches cannot fly'.

```python
# Write your code here for Q11
#Method overriding ka matlab hota he child class parent class ke kisi bhi method ko apni zaroorat ke mutabiq dobara define kar deti he .jab child object ka use hota he to child wala khud execute hota he . 

# Parent Class
class Bird:
    def fly(self):
        print("Most birds can fly")


# Child Class
class Ostrich(Bird):
    def fly(self):   # Method Overriding
        print("Ostriches cannot fly")


# Objects
bird = Bird()
ostrich = Ostrich()

bird.fly()
ostrich.fly()
```

**Q12:** Polymorphism function ke zariye kaise achieve hoti hai? Aik function banayein `make_sound(animal_object)` jo har tarah ke animal class ka sound method automatically call kare.

```python
# Write your code here for Q12
#polymorphism ka matlab he "one interface,many forms".Python se hum ek hi function bana sakte hein jo different classes ke objects ke sath kaam kar rahe.function automatically us object ka relevant method call kar leti he.

#parent class
class animal:
    def sound(self):
        print("Animal make a sound")

#child class1
class Dog(animal):
    def sound(self):
        print("Dog barks")

#child class2
class cat(animal):
    def sound(self):
        print("cat meows")

#polymorphism function 
def make_sound(animal_object) :
     animal_object.sound()


#object
dog = Dog()
cat = cat()
animal = animal()

make_sound(dog)
make_sound(cat)
make_sound(animal)
```

**Q13:** Operator Overloading kya hoti hai? Do objects ko `+` operator ke zariye add karne ke liye `__add__` magic method ka use kaise karte hain? Code likhein.

```python
# Write your code here for Q13
#Operator overloading ka matlab he ke hum python ke built in operators (+,-,*) ko apni class ke objects ke liye  ko apni marzi ke mutabiq define kar sakte hein.

class number:
    def __init__(self,value):
        self.value = value

    #overloading +operator
    def __add__(self,other):
        return number(self.value + other.value)

    def display(self):
        print("value",self.value)

#objects
num1 = number(10)
num2 = number(20)

#using + operators
result = num1 + num2

result.display()
```

**Q16:** `abc` module ka use karte hue aik abstract class `Computer` banayein jisme abstract method `process()` ho. Is class ka object banane ki koshish karein aur dekhein kya hota hai.

```python
# Write your code here for Q16
# abc (Abstract Base Class) module ka use abstract classes aur abstract methods banane ke liye kiya jata hai. Agar kisi class mein abstract method ho, to us class ka direct object create nahi kiya ja sakta. Pehle us abstract class ko inherit karke abstract methods implement karna zaroori hota hai.

from abc import ABC, abstractmethod

# Abstract Class
class Computer(ABC):

    @abstractmethod
    def process(self):
        pass

# Abstract class ka object banane ki koshish
c = Computer()
```

**Q17:** Q16 wali `Computer` class ko inherit karke do child classes `Laptop` aur `Desktop` banayein aur dono mein `process()` method ko mandatory implement karein.

```python
from abc import ABC, abstractmethod

# Abstract Class
class Computer(ABC):

    @abstractmethod
    def process(self):
        pass


# Child Class 1
class Laptop(Computer):

    def process(self):
        print("Laptop is processing data.")


# Child Class 2
class Desktop(Computer):

    def process(self):
        print("Desktop is processing data.")


# Objects
l1 = Laptop()
d1 = Desktop()

# Calling methods
l1.process()
d1.process()
```

**Q18:** Aik abstract class `Car` banayein jisme aik concrete (normal) method `start_engine()` ho aur aik abstract method `drive()` ho. Child class bana kar dono ko test karein.

```python
# Write your code here for Q18
from abc import ABC, abstractmethod

# Abstract Class
class Car(ABC):

    # Concrete (Normal) Method
    def start_engine(self):
        print("Engine started.")

    # Abstract Method
    @abstractmethod
    def drive(self):
        pass


# Child Class
class Toyota(Car):

    def drive(self):
        print("Toyota is driving on the road.")


# Object of Child Class
car = Toyota()

# Calling Methods
car.start_engine()
car.drive()
```

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
# Write your code here for Q20
from abc import ABC, abstractmethod

# Abstract Class
class Payment(ABC):

    @abstractmethod
    def pay(self, amount):
        pass


# Child Class 1
class CreditCardPayment(Payment):

    def pay(self, amount):
        print(f"Rs. {amount} paid using Credit Card.")


# Child Class 2
class UPIPayment(Payment):

    def pay(self, amount):
        print(f"Rs. {amount} paid using UPI.")


# Objects
credit = CreditCardPayment()
upi = UPIPayment()

# Payment
credit.pay(5000)
upi.pay(2500)
```
